**Task1**:

Describe the similarities and differences between an int and a float in python.

**Answer1**:

Both int and float represent numeric data types, but int holds whole numbers without a decimal point, while float stores numbers with decimal points.

**Task2**:

Describe the similarities and differences between: strings, lists, and dicts.

**Answer2**:

Strings hold text, lists hold ordered groups of items, and dictionaries hold pairs of keys and values, but only lists and dictionaries can be changed after they're created.

**Task3**:

What is a numpy array? Why would we ever use one over a list or dict?

**Answer3**:

A numpy array is a special type of list used for storing and working with numbers, and we use it instead of lists or dicts when we need faster math operations or handle lots of data.

**Task4**:

What is a pandas dataframe? Why would we ever want to use one over a numpy array?

**Answer4**:

A pandas dataframe is like a table used for organizing data in rows and columns, and we use it over a NumPy array when we need labels for rows and columns, or to easily query and filter data based on conditions.

**Task5**:

What are for loops, and why would we ever use them?

**Answer5**:

Because most of the time there is a dataframe way to find/filter data.
